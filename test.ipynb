<h2>Linear Plot (Examples for the tricks)</h2>
<h3>Create values</h3>

```python
# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
# Define a linear relationship: y = 2x + 1
y = 2 * x + 1
```

<h3>Plotting</h3>

```python
# Create the plot
plt.plot(x, y, label='y = 2x + 1')
# Setup plot aesthetics
mpl.rcParams['font.family'] = 'Arial'
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Relationship Plot')
plt.legend(frameon=False)
plt.text(1, 15, 'y = 2x + 1', fontsize=12, color='blue') #Optional annotation (preferably in blue color)
plt.grid(False)
plt.figure(figsize=(8,6), dpi=300)
# Display the plot
plt.show()
```
